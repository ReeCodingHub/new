```python
from nltk.tokenize import sent_tokenize,word_tokenize
corpus="""Hi, this is my 1st day of NLP.
I'm trying to master LLMs"""
print(corpus)
```

```python
nltk.download('punkt')
```

```python
#tokenization
#returns list of sentences divided by new line
len(sent_tokenize(corpus))
```

```python
words=word_tokenize(corpus)
```

```python
##JUST FOR FUN
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords=set(STOPWORDS)
comment_words=' '.join(words)+" "
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)

# plot the WordCloud image
plt.figure(figsize = (5, 5), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

```python
#Stemming
from nltk.stem import PorterStemmer,RegexpStemmer,SnowballStemmer
words=['gone','going','go','myself','viewer','fairly','congratulate','wishes']
stemming=PorterStemmer()
```

```python
for word in words:
  print(word+'---->'+stemming.stem(word))
'''disadv of Stemming
1.meaning of the word may change'''
```

```python
reg=RegexpStemmer('ing$|ed$')
for word in words:
  print(word+'---->'+reg.stem(word))
#Disadv
#Simply removes the exp given to regexp
```

```python
snw=SnowballStemmer("english") #available for multiple languages
for word in words:
  print(word+'---->'+snw.stem(word))
```

```python
#Conclusion: SnowballStemmer performs better than other stemming methods
```
